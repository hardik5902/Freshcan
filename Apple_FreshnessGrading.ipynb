```python
X2 = df1.drop(['BinaryClass','image_id','Grade','assigned_clusters'],axis=1)
y2 = df1['Grade']
```

```python
X2
```

```python
sc1 = StandardScaler()
X2 = sc1.fit_transform(X2)
```

```python
joblib.dump(sc1,'AppleScaler_2.save')
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(X2,y2,test_size=0.25,random_state=115426)
```

```python
clf2 = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric = None)
models,predictions = clf2.fit(x2_train,x2_test,y2_train,y2_test)
print(models)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x2_train,y2_train)
```

```python
ypred1_train = rfc.predict(x2_train)
ypred1_test = rfc.predict(x2_test)
```

```python
print(f1_score(y2_test,ypred1_test,average='weighted'))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 4, 8],
    'min_samples_leaf': [1, 2, 4],
}

# Initialize the classifier
clf = RandomForestClassifier(random_state=42)

# Use GridSearchCV to perform a search over the hyperparameter grid
grid_search = GridSearchCV(clf, param_grid, cv=5)
grid_search.fit(x2_train, y2_train)

# The best hyperparameters found by GridSearchCV
best_params = grid_search.best_params_

# Train the final classifier using the best hyperparameters found
best_clf = RandomForestClassifier(**best_params, random_state=42)
best_clf.fit(x2_train, y2_train)
```

```python
best_clf.score(x2_test,y2_test)
```

```python
joblib.dump(best_clf, "applemodel_2.joblib")
```
